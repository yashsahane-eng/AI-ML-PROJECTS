```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme(style="whitegrid")
```

```python
df=sns.load_dataset("tips")
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
sns.scatterplot(x="total_bill",y="tip",data=df)
```

```python
sns.barplot(x="day",y="tip",data=df)
plt.title("Average Tip By Day Of the Week ")
plt.show()
```

```python
sns.barplot(x="time",y="tip",hue="smoker",data=df)
plt.title("Average tip by time and smoker")
plt.show()
```

```python
sns.histplot(df['tip'],kde=True)
plt.title("Distribution of Tip")
plt.show()
```

```python
sns.boxplot(x="day",y="tip",data=df)
plt.title("Distribution of Tips By Day Of The Week ")
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```
